```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
import pandas as pd
data = {
    'Days' :['Mon','Tue','WED','Thurs','Fri'],
    'Study_Hours':[2,3,2.5,4,3.5],
    'Marks':[50,60,55,70,65],
    'Attendance':[60,70,65,80,75],
    'Sleep_Hours':[8,7,8,6,7],
    'Assignment_Completed':[1,2,1,3,2]
}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
#1 Line CHART (TREND)
plt.plot(df['Days'], df['Study_Hours'], marker='*')
plt.title('Study hours over days')
plt.xlabel('Days')
plt.ylabel('Study hours')
plt.show()
```

```python
from numpy import ma
#Advanced line chart
plt.figure(figsize=(7,4))


plt.plot(df['Days'], df['Study_Hours'], marker='o', label='Study_hours'
,color ='Blue')

plt.plot(df['Days'],df['Marks'],marker ='s',color='Red',label='Marks')

plt.title('Amod')
plt.xlabel('Days')
plt.ylabel('Marks')

plt.legend()
plt.show()
```

```python
#2. Bar Chart (comparison)
plt.bar(df['Days'], df['Marks'],color='green')
plt.title('Amod')
plt.xlabel('Days')
plt.ylabel('Marks')
plt.show()
```

```python
import matplotlib.pyplot as plt
# Advanced Bar graph
bars = plt.bar(df['Days'], df['Marks'],color='red')
for bar in bars:
    y = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, y, str(y), ha='center', va='bottom')
    plt.title('Amod')
    plt.xlabel('Days')
    plt.ylabel('Marks')
```

```python
# 3. Histogram (Distribution)
# shows how values are spread

data = {
    'Marks':[50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
}
df = pd.DataFrame(data)
plt.hist(df['Marks'],bins=3, edgecolor='black',alpha=0.30)
plt.title('Amod')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.show
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Example data (more values for better histogram)
marks = [50,60,55,70,65,75,80,85,90,45,68,72,77,82,88]
# Calculate mean\
mean_value = np.mean(marks)
plt.figure(figsize=(7,4))
plt.hist(marks,bins=6,color='blue',linestyle='--',linewidth=2,label='Mean')
#mean line
plt.axvline(mean_value,color='red',linestyle='--',linewidth=2,label='Mean')
plt.title('Amod')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.legend()

plt.grid(axis='y',linestyle='--',alpha=0.5)
plt.show()
```

```python
# 4. scatter plot(relationship)
import matplotlib.pyplot as plt
plt.scatter(df['Study_Hours'],df['Marks'],color='green')
plt.title('Amod')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
df['Result']=['Fail','Pass','Fail','Pass','Pass']
colors = ['red' if result == 'Fail' else 'green' for result in df['Result']]
plt.scatter(df['Study_Hours'],df['Marks'],color=colors)
plt.title('Amod')
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.show()
```

```python
data = {
    'Day': ['Mon','Tue','Wed','Thu','Fri','Sat','Sun','Mon','Tue','Wed'],
    'Region': ['North','South','East','West','North','South','East','West','North','South'],
    'Sales': [200,220,250,270,300,350,400,280,260,290],
    'Profit': [20,25,30,35,50,60,80,40,38,45],
    'Customers': [30,35,40,45,50,65,80,55,48,52],
    'Category': ['Food','Clothing','Electronics','Food','Clothing','Electronics','Food',
                 'Clothing','Electronics','Food']
}

df = pd.DataFrame(data)
print(df)
```

```python
# Line Plot
sns.lineplot(x='Day',y='Sales',data = df)
plt.title('Amod')
plt.show()
```

```python
# Bar Plot
sns.barplot(x='Day',y='Profit',data=df)
plt.title('Amod')
plt.show()
```

```python
# Histogram
sns.histplot(df['Sales'],bins=5)
plt.title('Amod')
plt.show()
```

```python
#Scatter plot
sns.scatterplot(x='Sales',y='Profit',data=df)
plt.title('Amod')
plt.show()
```
